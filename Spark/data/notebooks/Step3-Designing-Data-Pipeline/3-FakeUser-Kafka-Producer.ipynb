```python
import time
import requests
import re
from pprint import pformat
from json import dumps
from kafka import KafkaProducer
import random
```

```python
producer = KafkaProducer(bootstrap_servers=['kafka-broker:29092'], value_serializer=lambda x: dumps(x).encode('utf-8'),
                        key_serializer=str.encode )
TOPIC_NAME="User-Raw-Data"
```

```python
url = "https://randomuser.me/api?nat=IR"
delay = 5
cnt=0
while True:
    try : 
        user_counts=random.randint(20,50)
        print( f"{url}&results={user_counts}")
        response = requests.request('GET', f"{url}&results={user_counts}", headers={'User-Agent': 'Chrome/81'})
        if response.status_code == requests.codes.ok:
            items = response.json()['results']
            for item in items:
                cnt+=1
#                 print("-=-"*20)
                print(f"#{cnt:4}")
#                 print(f"Username : {item['login']['username']}\n  {pformat(item['name'])}")
                producer.send(TOPIC_NAME, value=item, key= item['login']['username'])
        else:
            print("Error in fetch data: {err}".format(err=response.status_code))
        time.sleep(delay)
    except Exception as e:
        print(e)
        continue

```
